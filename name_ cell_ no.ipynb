```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

Name = []
Contact_No = []
url = 'https://www.ibccoaching.com.br/coaches-formados/'
source = requests.get(url).text
soup = BeautifulSoup(source, 'html.parser')

# Find all the coach items
coach_items = soup.find_all('li', class_='kleo-masonry-item')

profile_links = []

# Iterate over each coach item
for coach_item in coach_items:
    # Extract the coach's profile link
    profile_link = coach_item.find('div', class_='item-title').a['href']
    
    # Append the profile link to the list
    profile_links.append(profile_link)

# Iterate over the profile links
for profile_link in profile_links:
    source = requests.get(profile_link).text
    soup = BeautifulSoup(source, 'html.parser')

    # Continue with further processing or scraping for each profile
    
    # Example: Extract the coach's name
    name = soup.find('h1', class_='page-title').text.strip()
    cont = soup.find('dl', class_='field_celular')
    Cell_no = cont.find('dd').find('p').text.strip()


    Name.append(name)
    Contact_No.append(Cell_no)


df=pd.DataFrame({
    'Name':Name,
    'Contact_No':Contact_No
})
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ibccoaching.com.br/coaches-formados/'

def get_members_data(page_url):
    source = requests.get(page_url).text
    soup = BeautifulSoup(source, 'html.parser')

    member_items = soup.find_all('li', class_='kleo-masonry-item')

    members_data = []

    for member_item in member_items:
        name = member_item.find('div', class_='item-title').text.strip()
        profile_link = member_item.find('div', class_='item-title').a['href']
        
        current_url = url + profile_link
        next_source = requests.get(current_url).text
        soup_2 = BeautifulSoup(next_source, 'html.parser')
        phone_element = soup_2.find('dl', class_='field_celular')

        if phone_element:
            phone_number = phone_element.find('dd').text.strip()
        else:
            phone_number = ''  # Assign a default value if phone_element is None

        members_data.append((name, phone_number))

    return members_data


all_members_data = []

# Start scraping from the first page
page_number = 1
current_url = url + '?upage=' + str(page_number)

# Scrape data from each page until there are no more pages
while True:
    members_data = get_members_data(current_url)
    all_members_data.extend(members_data)
    
    # Move to the next page
    page_number += 1
    current_url = url + '?upage=' + str(page_number)

    source = requests.get(current_url).text
    soup = BeautifulSoup(source, 'html.parser')
    pagination_div = soup.find('div', class_='pagination-links')

    # Check if there is a next page
    if not pagination_div or 'next' not in pagination_div.text.lower():
        break

# Print the scraped data
for i, data in enumerate(all_members_data, 1):
    print("Member", i)
    print("Name:", data[0])
    print("Cell Number:", data[1])
    print()
```
